```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Я выбрал экспоненциальное распределение 

```python
expon_rv = sts.expon ()
```

```python
sample = expon_rv.rvs (1000)
x = np.linspace(0,7,1000)
pdf = expon_rv.pdf(x)
plt.plot (x,pdf,label='Theoretical PDF')
plt.hist (sample, normed = True,label='Hist RVS')
plt.legend(loc='upper right')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

Для удобства напишем функцию выборочного среднего(samples_mean) при разных обьемах выборок (n) и количеством выборок = 1000

```python
def samples_mean (n):
    a = []
    for i in range (1000):
        xS = sum(expon_rv.rvs(n))/n
        a.append (xS)
    return a
```

Теперь построим гистрограммы для разных n (5,10,50)

```python
#Функции нормального распределения (параметры посчитаны под гистрограммами)
norm_rv5 = sts.norm(1, 0.45)
norm_rv10 = sts.norm(1, 0.32)
norm_rv50 = sts.norm(1, 0.14)

x = np.linspace (0,3,1000)
plt.plot (x,norm_rv5.pdf(x),label='theoretical pdf')
plt.hist (samples_mean (5), normed = True,label='n = 5')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
plt.plot (x,norm_rv10.pdf(x),label='theoretical pdf')
plt.hist (samples_mean (10), normed = True,label='n = 10', color = 'yellow')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
plt.plot (x,norm_rv50.pdf(x),label='theoretical pdf')
plt.hist (samples_mean (50), normed = True,label='n = 50',color = 'red')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

Теоретические среднее значение и дисперсия у данного экспоненциального распределение будут равны 1 (m = 1, D = 1).

Соответственно дисперсии для каждой из выборок будут равны:
При n = 5, D = 0.2 (Sigma = 0.45, m = 1), 
При n = 10, D = 0.1 (Sigma = 0.32, m = 1),
При n = 50, D = 0.02 (Sigma = 0.14, m = 1).

# Выводы

При увеличении обьема выборочного среднего (n), значения все больше концентрируются вокруг среднего значения, а так  же увеличивается точность апроксимации нормальным распределением
