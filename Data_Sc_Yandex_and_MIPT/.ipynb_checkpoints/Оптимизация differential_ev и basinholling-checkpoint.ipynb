```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange (0,16,0.2)
def f(x):
    return np.sin(x/5.0)*np.exp(x/10.0)+ 5*np.exp(-x/2.0)
plt.plot(x,f(x))
plt.show()
```

```python
A1 = [[1 ** n for n in range (0,2)], [15 ** n for n in range (0,2)]]
B1 = [[f(1)],[f(15)]]
x1 = np.linalg.solve(A1, B1)
print A1
print B1
print x1
```

```python
def f1(x):
    return x1[0] + x1[1]*x
```

```python
plt.plot (x,f1(x))
plt.show
```

```python
A2 = [[1**n for n in range (0,3)],[8**n for n in range (0,3)],[15 ** n for n in range (0,3)]]
B2 = [[f(1)],[f(8)],[f(15)]]
x2 = np.linalg.solve (A2,B2)
```

```python
def f2(x):
    return x2[0] + x2[1]*x + x2[2]*x**2
plt.plot(x, f2 (x))
plt.show
```

```python
A3 = [[1**n for n in range (0,4)],[4**n for n in range (0,4)],[10**n for n in range (0,4)],[15 ** n for n in range (0,4)]]
B3 = [[f(1)],[f(4)],[f(10)],[f(15)]]
x3 = np.linalg.solve (A3,B3)
print x3
```

```python
def f3(x):
    return x3[0] + x3[1]*x + x3[2] *x**2 + x3[3]*x**3
plt.plot (x,f3(x))
plt.show
print f3(1)
```
